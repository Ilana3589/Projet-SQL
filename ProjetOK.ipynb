## Merging the Datasets

Since the data is spread across three separate files (one-star, two-star, and three-star restaurants), we first merge them into a single dataset. This allows for easier analysis and comparison.

```python
df1 = pd.read_csv('one-star-michelin-restaurants.csv')
df2 = pd.read_csv('two-stars-michelin-restaurants.csv')
df3 = pd.read_csv('three-stars-michelin-restaurants.csv')
```

I add a new column ("Stars") to each dataset to indicate how many stars each restaurant has.

```python
df1["Stars"] = 1
```

```python
df2["Stars"] = 2
```

```python
df3["Stars"] = 3
```

I concatenate the three datasets into one using

```python
df_michelin = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
df_michelin.head()
```

## Handling Missing Values

Some data points in the dataset are missing . We remove rows where too many values are missing

Fill missing zip codes with ''unknown''

```python
print(df_michelin.isnull().sum())
```

I remove rows with more than 2 missing values

```python
df_cleaned = df_michelin.dropna(thresh=len(df_michelin.columns) - 2)
```

I fill missing zip codes with "Unknown

```python
df_cleaned["zipCode"] = df_cleaned["zipCode"].fillna("Unknown")
```

```python
print(df_cleaned.isnull().sum())
```

I Replace Missing Prices with "Unknown"

```python
df_cleaned["price"] = df_cleaned["price"].fillna("Unknown")
```

```python
print(df_cleaned.isnull().sum())
```

## Standardizing Text Data

Text data, like cuisine types, may have inconsistencies in capitalization and spaces

We ensure uniformity by converting all text to lowercase.

```python
df_cleaned["cuisine"] = df_cleaned["cuisine"].str.lower().str.strip()
```

```python
df_cleaned["cuisine"].unique()
```

## Removing Outliers

Some records may contain incorrect geographic coordinates (e.g., latitude 9999). To maintain data accuracy, we filter out invalid values.

I Keep only records where:
latitude is between -90 and 90
longitude is between -180 and 180

```python
df_cleaned = df_cleaned[
    (df_cleaned["latitude"].between(-90, 90)) & 
    (df_cleaned["longitude"].between(-180, 180))
]
```

```python
df_cleaned.head()
```

After all these cleaning steps, the dataset is ready for analysis.
It now contains accurate, consistent, and structured data without missing values or outliers.

I want to create a grouped bar chart to show the predominant cuisine type for each Michelin star level


```python
import matplotlib.pyplot as plt

cuisine_star_counts = df_michelin.groupby(["cuisine", "Stars"]).size().unstack(fill_value=0)

threshold = 10  
filtered_cuisine_star_counts = cuisine_star_counts.loc[(cuisine_star_counts >= threshold).any(axis=1)]

plt.figure(figsize=(12, 6))
filtered_cuisine_star_counts.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")

plt.title("Michelin Star Cuisine Distribution (Filtered Stacked Bar Chart)", fontsize=14)
plt.xlabel("Cuisine Type", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.legend(title="Michelin Stars")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

star_counts = df_michelin["Stars"].value_counts()

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, star in enumerate(sorted(df_michelin["Stars"].unique())):
    cuisine_counts = df_michelin[df_michelin["Stars"] == star]["cuisine"].value_counts()
    axes[i].pie(cuisine_counts.head(10), labels=cuisine_counts.index[:10], autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f"{star}-Star Restaurants")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_cuisines_per_star = df_michelin.groupby("Stars")["cuisine"].value_counts().groupby(level=0).head(5).unstack(fill_value=0)

plt.figure(figsize=(12, 6))
top_cuisines_per_star.plot(kind="bar", figsize=(12, 6), colormap="viridis")

plt.title("Predominant Cuisine Type per Michelin Star Level (Grouped Bar Chart)", fontsize=14)
plt.xlabel("Michelin Stars", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.legend(title="Cuisine Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

The bar chart clearly illustrates the predominant cuisine types for each Michelin star level (1, 2, and 3 stars)

Across all Michelin star levels, contemporary and modern cuisine appear to be the most represented

1-star restaurants showcase a wider variety of cuisine types, including market cuisine, classic cuisine, and contemporary cuisine.

In the 2-star category, modern and creative cuisine remain dominant

Cantonese and Contemporary cuisine are the most predominant types among 3-star Michelin restaurants. Contemporary cuisine, often a fusion of various culinary traditions, is  highly favored at the 3-star Michelin level.Many contemporary restaurants incorporate Asian, European, and Latin flavors, blending ingredients and cooking methods from different regions.
This makes contemporary cuisine highly adaptable, attracting a global audience. This trend reflects key characteristics of high-end dining and Michelin's evaluation criteria.They reflect a balance of heritage and creativity

## Analysis of Price Distribution Across Michelin Star Levels

Here is the stacked bar chart showing the price distribution across Michelin star levels, now with clear text labels instead of dollar signs.
Each Michelin star category (1, 2, 3 stars) is represented on the x-axis.
The different colors represent price levels, showing how pricing varies between Michelin star levels.

```python
# Replace dollar signs with text labels to avoid  conflicts
price_labels = {
    "$": "Very Cheap",
    "$$": "Cheap",
    "$$$": "Moderate",
    "$$$$": "Expensive",
    "$$$$$": "Luxury"
}

# Apply new labels
df_michelin["price_category"] = df_michelin["price"].map(price_labels)

# Group by Michelin stars and price category
price_distribution_fixed = df_michelin.groupby("Stars")["price_category"].value_counts().unstack().fillna(0)

# Create the stacked bar chart again with corrected labels
plt.figure(figsize=(12, 15))
price_distribution_fixed.plot(kind="bar", stacked=True, figsize=(12, 15), cmap="plasma")

# Labels and title
plt.title("Price Distribution Across Michelin Star Levels", fontsize=14)
plt.xlabel("Michelin Star Level", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.legend(title="Price Level", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

1-star restaurants have the widest range of price levels, from cheap  to luxury 
2-star restaurants start to concentrate more in the moderate to expensive range.
3-star restaurants are overwhelmingly in the luxury  category, with very few in the lower price ranges.

A significant portion of 1-star restaurants fall in the cheap and moderate price range

This suggests that a Michelin star does not always mean ultra-expensive dining—many 1-star restaurants maintain more accessible pricing.

The vast majority of 3-star restaurants fall into the luxury category, with very few exceptions

This analysis confirms that while Michelin stars reward culinary excellence, the cost of dining increases significantly with higher ratings due to the demand for superior ingredients, service, and presentation.
